```python
from dotenv import load_dotenv
import os
import base64
import requests
import json

id = ""
json_result = ""
```

```python
#env files contain credentials in key-value format for services used by the program they're building. 
# They're meant to be stored locally and not be uploaded to code repositories online for everyone to 
# read.

# Here the env stores my client id and client secret

#loads the creds from the env file
load_dotenv()

#accesses each credential and puts it into corresponding variable
client_id = os.getenv("CLIENT_ID")
client_secret = os.getenv("CLIENT_SECRET")

#function that automates getting a token for us

# Note: here we are using the client credentials auth method
# and recieving the access token. This eliminates the user (us)
# from granting any authorization in the process. As a result,
# we can only access non-user/non-personal data (no user data essentially)
# only the server (spotify) grants access and communicates with
# with the client/app we made which is requesting data.

# You will need the access token everytime you make an API
# call/request. The token expires after an hour.

# here, the auth is done by sending a POST request (using the requests
# library since we are working with python and requests is a package
# that allows you to make HTTP requests (with diff methods, POST being
# one) with Python)
# you make the POST request to the /api/token endpoint or url with
# the following parameters:
# 
# grant_type which is client_credentials
# 
# header parameter:
# Authorisation: base 64 encoding of the client id and secret 
# content-type: application/x-www-form-urlencoded
# 
# 
# 
# 
# 
# 
# 


def get_token():
    
    #
    auth_string = client_id + ":" + client_secret
    auth_bytes = auth_string.encode("utf-8")
    auth_base64 = str(base64.b64encode(auth_bytes), "utf-8")

    url = "https://accounts.spotify.com/api/token"
    
    headers = {
        "Authorization": "Basic " + auth_base64,
        "Content-Type": "application/x-www-form-urlencoded"
    }
    
    data = {"grant_type": "client_credentials"}
    
    #finally making the post request after setting up above all required parameters
    result = requests.post(url, headers= headers, data= data)
    
    #converts resulting json string data containing the access token
    #into a dictionary that we can access
    json_result = json.loads(result.content)
    
    #token var stores the value associated with the key "access_token" 
    #inside the json-converted-dictionary
    token = json_result["access_token"]
    
    return token

token = get_token()

    
```

```python
def get_auth_header(token):
    return {"Authorization": "Bearer " + token}
```

```python
# valence
# number [float]
# A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. 
# Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), 
# while tracks with low valence sound more negative (e.g. sad, depressed, angry).

# Example value: 0.428
# Range: 0 - 1

# I will use this valence value to determine the mood. I will print out all the 
# top 50 tracks' valence now

valences = []
for x in features['audio_features']:
    print(x['valence'])
    valences.append(x['valence'])
```
